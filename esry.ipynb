```python
import pytesseract
import cv2
import pandas as pd
from PIL import Image
from google.colab import files
```

```python
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
img = cv2.imread(image_path)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 3)

text = pytesseract.image_to_string(gray)
print("النص المُستخرج:\n", text)
```

```python
db_file = files.upload()
db_path = list(db_file.keys())[0]
df = pd.read_excel(db_path)
```

```python
found = False
for i, row in df.iterrows():ج
    if row['اسم الدواء'] in text and row['رقم التشغيلة'] in text:
        found = True
        print(f"\n✅ الدواء مطابق! بيانات التشغيلة سليمة:\n{row}")
        break

if not found:
    print("\n⚠️ تحذير: الدواء غير مطابق لقاعدة البيانات – احتمال يكون مزيف أو غير مسجل.")
```

```python
from pyzbar.pyzbar import decode
from PIL import Image

# رفع صورة فيها باركود
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# فتح الصورة
img = Image.open(image_path)

# قراءة الباركود
barcodes = decode(img)

if not barcodes:
    print("⚠️ لم يتم العثور على باركود.")
else:
    for barcode in barcodes:
        barcode_data = barcode.data.decode("utf-8")
        print(f"✅ تم قراءة الباركود: {barcode_data}")
```

```python
matched_row = df[df["باركود"] == barcode_data]

if not matched_row.empty:
    print("✅ الدواء موجود في قاعدة البيانات:\n", matched_row)
else:
    print("⚠️ الباركود غير موجود – الدواء غير مسجل أو مزيف.")
```
